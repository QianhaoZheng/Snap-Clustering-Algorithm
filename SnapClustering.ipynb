数据预处理

```python
df1 = df_update[['Travo ID','Hotel Booking Rank','Hotel Impression Rank','Availability','Win Rate','Weighted Delta to Winning Rate']]
df1
```

```python
df2 = df[['Hotels Travolutionary ID','Hotels Hotel Bookings Rank','Hotels Hotel Impressions Rank','Rate Share Availability','Rate Share Win Rate','Rate Share Weighted Delta to Winning Rate']]
rename_map = {'Hotels Travolutionary ID':'Travo ID','Hotels Hotel Bookings Rank': 'Hotel Booking Rank','Hotels Hotel Impressions Rank': 'Hotel Impression Rank', 'Rate Share Availability':'Availability', 'Rate Share Win Rate':'Win Rate','Rate Share Weighted Delta to Winning Rate': 'Weighted Delta to Winning Rate'}
df2 = df2.rename(columns=rename_map)
```

```python
df2
```

```python
df2['Hotel Booking Rank'] = df2['Hotel Booking Rank'].str.replace(',', '').astype(int)
df2['Hotel Impression Rank'] = df2['Hotel Impression Rank'].str.replace(',', '')
df2['Hotel Impression Rank'] = df2['Hotel Impression Rank'].fillna(0)
df2['Hotel Impression Rank'] = df2['Hotel Impression Rank'].astype(int)
df2['Availability'] = df2['Availability'].str.rstrip('%')
df2['Availability'] = df2['Availability'].astype(float)
df2['Availability'] = df2['Availability'] / 100
df2['Win Rate'] = df2['Win Rate'].str.rstrip('%')
df2['Win Rate'] = df2['Win Rate'].astype(float)
df2['Win Rate'] = df2['Win Rate'] / 100
df2['Weighted Delta to Winning Rate'] = df2['Weighted Delta to Winning Rate'].str.rstrip('%')
df2['Weighted Delta to Winning Rate'] = df2['Weighted Delta to Winning Rate'].astype(float)
df2['Weighted Delta to Winning Rate'] = df2['Weighted Delta to Winning Rate'] / 100
```

```python
df1['Travo ID'] = df1['Travo ID'].replace('Not Mapped', np.nan)
df2['Travo ID'] = df2['Travo ID'].replace('Not Mapped', np.nan)
df1['Travo ID'] = df1['Travo ID'].astype(float).astype(pd.Int64Dtype())
df2['Travo ID'] = df2['Travo ID'].astype(float).astype(pd.Int64Dtype())
df1 = df1.dropna(subset=['Travo ID'])
df2 = df2.dropna(subset=['Travo ID'])
```

```python
df2
```

```python
df2_processed = df2[~df2['Travo ID'].isin(df1['Travo ID'])]
final_df = pd.concat([df1, df2_processed], ignore_index=True)
final_df
```

聚类分析选阈值

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import pandas as pd

# 仅选择最后三列进行聚类分析
selected_data = final_df[['Availability', 'Win Rate', 'Weighted Delta to Winning Rate']]

# 删除包含NaN值的行
selected_data = selected_data.dropna()

# 数据预处理
scaler = StandardScaler()
scaled_data = scaler.fit_transform(selected_data)

# 聚类分析
kmeans = KMeans(n_clusters=2, random_state=42)  # 假设我们将每个指标分为两个群集
kmeans.fit(scaled_data)

# 获取每个指标的阈值
centroids = scaler.inverse_transform(kmeans.cluster_centers_)  # 将聚类中心的标准化数值转换为原始数值
thresholds = pd.Series(centroids.mean(axis=0), index=selected_data.columns)  # 计算每个指标的阈值

print("Thresholds for each metric:")
print(thresholds)
```

```python
import seaborn as sns
fig, axes = plt.subplots(1, 3, figsize=(15, 5))
for i, column in enumerate(selected_data.columns):
    sns.histplot(selected_data[column], kde=True, ax=axes[i])
    axes[i].axvline(x=thresholds[i], color='r', linestyle='--', label=f'Threshold: {thresholds[i]:.2f}')
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].legend()

plt.show()
```
